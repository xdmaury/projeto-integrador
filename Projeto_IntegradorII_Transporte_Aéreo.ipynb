# 3. Quais grupos de voos mais transportam cargas pagas?

```python
#Filtrando dados em 5 anos
df_carga = df_go.query("ANO >= 2017 and ANO < 2022")
```

```python
df_carga = df_carga[['ANO','MES','GRUPO_DE_VOO', 'CARGA_PAGA_KM']].groupby(['ANO', 'MES', 'GRUPO_DE_VOO']).sum().reset_index()
```

```python
df_carga = df_carga.drop(df_carga[df_carga['GRUPO_DE_VOO']=='IMPRODUTIVO'].index)
```

```python
df_carga = df_carga.groupby(['ANO', 'GRUPO_DE_VOO']).sum().drop(columns='MES').reset_index()
```

```python
df_carga = df_carga
```

```python
import plotly.express as px

fig = px.bar(df_carga, x="ANO", y="CARGA_PAGA_KM",
             color='GRUPO_DE_VOO', barmode='group',
             height=600)
fig.show()
```

Nos ultimos 5 anos as 3 maiores empresas que transpotaram cargas pagas to Grupo de voos não regular

```python
df_carga_area = df_go[['ANO', 'MES', 'EMPRESA_NOME','GRUPO_DE_VOO', 'CARGA_PAGA_KM']].query("ANO >= 2017 and ANO < 2022 and GRUPO_DE_VOO =='NÃO REGULAR'").groupby(['ANO', 'EMPRESA_NOME']).sum().reset_index()
```

```python
anos = list(dict.fromkeys(df_carga_area['ANO'].values))
```

```python
# Obtendo as 3 empresas com mais cargas pagas por KM ao ano
lista = pd.DataFrame()
for ano in anos:
  maiores = df_carga_area.query(f'ANO == {ano}').sort_values(by=['CARGA_PAGA_KM'], ascending=False).head(3)
  lista = pd.concat([lista, maiores])

top3ciabyano = lista[['ANO', 'EMPRESA_NOME', 'CARGA_PAGA_KM']]
```

```python
import plotly.express as px

fig = px.bar(top3ciabyano, x="ANO", y="CARGA_PAGA_KM",
             color='EMPRESA_NOME', barmode='group',
             height=600)
fig.show()
```

# 8. A ocupação média dos voos tem relação com a quantidade de companhias aéreas em operação?

```python
#Filtrando dados em 10 anos 
df_ocupacao = df_go.query("ANO > 2009 and ANO < 2020 and PASSAGEIROS_PAGOS != 0")
```

```python
df_ocupacao['MEAN_OCUPACAO'] = (df_ocupacao['PASSAGEIROS_PAGOS'] + df_ocupacao['PASSAGEIROS_GRATIS'])/ df_ocupacao['ASSENTOS']
```

```python
df_ocupacao
```

```python
df_ocupacao[['ANO','QTD_CIA','MEAN_OCUPACAO']].corr()
```

# 9. Existe correlação entre a companha aérea com mais voos e a quantidade de passageiros transportados?

```python
#Filtrando dados em 10 anos 
df_filter = df_go.query("ANO > 2009 and ANO < 2020")
```

```python
df_cia = df_filter[[ "ANO", "MES",  "EMPRESA_SIGLA"]].groupby(['ANO', 'EMPRESA_SIGLA']).count().reset_index()
```

```python
df_qtd_passageiro = df_filter[[ "ANO", "MES",  "EMPRESA_SIGLA", "PASSAGEIROS_PAGOS", "PASSAGEIROS_GRATIS"]].groupby(['ANO', 'EMPRESA_SIGLA']).sum().reset_index()
df_qtd_passageiro
```

```python
df_qtd_passageiro = df_qtd_passageiro.groupby(['ANO']).sum().reset_index().drop(columns=['MES'])
```

```python
anos = list(dict.fromkeys(df_cia['ANO'].values))
```

```python
lista = []
for ano in anos:
  res = df_cia.query(f'ANO == {ano}')
  index = res['MES'].idxmax()
  lista.append(dict(res.loc[index]))

df_cia = pd.DataFrame(lista)
```

```python
df_cia.rename(columns={'MES': 'QTD_VOOS'}, inplace=True)
```

```python
df_cia
```

```python
df_cia['TOTAL_PASSAGEIROS_BY_ANO'] = df_qtd_passageiro['PASSAGEIROS_PAGOS'] + df_qtd_passageiro['PASSAGEIROS_GRATIS']
df_cia
```

O fato da companhia fazer muitos **Voos** é determinante para a quantidade de passageiros transportados?

```python
df_cia.corr()
```
